```python
colors_list = ['pink', 'green','red','blue','yellow']


df_value_top5['value'].plot(kind='pie',
                           figsize=(8,10),
                           autopct='%1.1f%%', 
                           startangle=90, 
                           shadow=False,
                           colors = colors_list)
                           
plt.title('Persentase top 5 area kepolisian dengan tingkat kejahatan tertinggi')
plt.axis('equal')
plt.show()
```

## WordClouds

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
df_london = df_london.reset_index()
```

```python
df_london
```

```python
df_london_crime = df_london.set_index('major_category')
```

```python
df_london_crime
```

```python
values_total = df_london_crime['value'].sum()
values_total
```

```python
max_words = 100
major_crime_string = ''
for major_category in df_london_crime.index.values:
    if len(major_category.split(' '))==1:
        repeat_num_times = int(df_london_crime.loc[major_category, 'value']/float(values_total)*max_words)
        major_crime_string = major_crime_string +((major_category + ' ') * repeat_num_times)
```

```python
df_london
```

```python
max_words = 100000
```

```python
major_crime_text = ', '.join(major_category for major_category in df_london['major_category'][:max_words])
```

```python
major_crime_wc = WordCloud(width=800, height=400, background_color="white").generate(major_crime_text)
```

```python
plt.figure(figsize=(20, 10))
plt.imshow(major_crime_wc, interpolation='bilinear') ## interpolasi untuk pixel gambar lebih baik
plt.axis("off")
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
london_mask = np.array(Image.open('london_map.png'))
```

```python
plt.imshow(london_mask,cmap=plt.cm.gray)
plt.axis('off')
```

```python
major_crime_wc = WordCloud(background_color="white",mask=london_mask)
major_crime_wc.generate(major_crime_text)
plt.figure(figsize=(10,20))
plt.imshow(major_crime_wc) ## interpolasi untuk pixel gambar lebih baik
plt.axis("off")
plt.show()
```

Dari WordCloud di atas diambil 100.000 kata pada Major Category Crime, dapat dilihat bahwa Criminal Damage menampilkan font yang besar dan warna yang mencolok sehingga Criminal Damage adalah kategori kejahatan yang paling sering terjadi di London
