## Hypothesis Testing

melakukan hypothesis testing untuk menentukan apakah nilai mean SALE PRICE signifikan berbeda dengan nilai tertentu atau tidak. 

Misalnya, apakah nilai mean SALE PRICE pada daerah tertentu signifikan lebih tinggi dari nilai rata-rata SALE PRICE pada seluruh data?

```python
df_neigh = df
```

```python
df_neigh.loc[df_neigh['SALE PRICE'] ==' -  ' , 'SALE PRICE'] = np.nan
```

```python
df_neigh['SALE PRICE']
```

```python
df_neigh['SALE PRICE'] = df_neigh['SALE PRICE'].astype(float)
```

```python
df_neigh['SALE PRICE'].info()
```

```python
from scipy import stats
# melakukan t-test dengan menggunakan nilai mean dari seluruh data sebagai nilai null hypothesis
t_stat, p_value = stats.ttest_1samp(df_neigh['SALE PRICE'], df_neigh['SALE PRICE'].mean())

print("T-Statistic: ", t_stat)
print("P-Value: ", p_value)
```

```python
df_neigh['SALE PRICE'] = df_neigh['SALE PRICE'].fillna(df['SALE PRICE'].mean())
```

```python
df_neigh['SALE PRICE']
```

```python
from scipy import stats
# melakukan t-test dengan menggunakan nilai mean dari seluruh data sebagai nilai null hypothesis
t_stat, p_value = stats.ttest_1samp(df_neigh['SALE PRICE'], df_neigh['SALE PRICE'].mean())

print("T-Statistic: ", t_stat)
print("P-Value: ", p_value)
```

T-Statistic sangat kecil, yaitu sekitar 2.370910898410455e-11, sehingga menunjukkan bahwa sample mean tidak jauh berbeda dari populasi mean.

P-Value sangat besar, yaitu sekitar 0.9999999999810829, sehingga menunjukkan bahwa sampel yang diamati diambil dari populasi yang memiliki rata-rata yang sama.
 
kesimpulannya, hasil ini menunjukkan bahwa tidak ada perbedaan signifikan antara nilai rata-rata populasi dan sampel pada tingkat signifikansi tertentu.
