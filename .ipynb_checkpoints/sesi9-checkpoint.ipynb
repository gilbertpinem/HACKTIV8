## Standar Deviasi

```python
## Library Scipy
x = [1,2,3,6,7,8,30,40,50]
std_dev = statistics.stdev(x)
print("Standar Deviasi adalah : ",std_dev)
variance = statistics.variance(x)
print("Variance adalah : ",variance)
```

```python
## Manual
x = [1,2,3,6,7,8,30,40,50]
n= len(x)
mean = sum(x)/n
var_ = sum((i-mean)**2 for i in x) /(n-1) ## Variance

std = var_**(1/2) #  Standar deviasi
print(var_)
print(std)
```

```python
stdev = np.sqrt(var_)
```

```python
stdev
```

```python
stdev_np = np.std(x, ddof=1)
```

```python
stdev_np
```
